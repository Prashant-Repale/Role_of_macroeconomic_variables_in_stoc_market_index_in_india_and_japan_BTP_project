# 1.Unit Root Test 

```python
from statsmodels.tsa.stattools import adfuller
import numpy as np
```

#### NIFTIY 50

```python
# 'c' represents “Intercept” 
print('"intecept is present"')
nse_c = adfuller(new_india_data['log(NSE)'],regression='c',autolag='AIC')
print('Test Statistic',round(nse_c[0],3))
print('p-value',round(nse_c[1],3))
print('Critical Values')
for key, value in nse_c[4].items():
    print(f'     {key} : {round(value,3)}')
print('  ')

# 'ct' represents “Trend and Intercept”
print('"Trend and Intercept both are present"')
nse_c = adfuller(new_india_data['log(NSE)'],regression='ct',autolag='AIC')
print('Test Statistic',round(nse_c[0],3))
print('p-value',round(nse_c[1],3))
print('Critical Values')
for key, value in nse_c[4].items():
    print(f'     {key} : {round(value,3)}')

```

 # 2.Multivariate Cointegration Test 

```python
from statsmodels.tsa.api import VAR
```

```python
model = VAR(new_india_data.iloc[:,1:])     # mushje ayha dataframe dena tha nahi to 2d array 
lag_results = model.select_order(maxlags=10)
print(lag_results.summary())
```

```python
print(lag_results)
```

```python
lag_results.selected_orders
```

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
```

```python
coin = coint_johansen(new_india_data.iloc[:,1:],det_order=1,k_ar_diff=1)
```

```python
print("Eigenvalues:", coin.eig)
print("Trace Statistic:", coin.lr1)
print("Trace Critical Values (90%, 95%, 99%):")
print(coin.cvt)

print("Max-Eigen Statistic:", coin.lr2)
print("Max-Eigen Critical Values (90%, 95%, 99%):")
print(coin.cvm)
```

```python
print(dir(coin))
```

```python
beta = coin.evec
print(np.round(beta,1))
```

```python
print(pd.DataFrame(coin.evec, 
                   index=["IR %", "log(NSE)", "log(m3)", "log(iip)", "log(cop)"], 
                   columns=['Vec1', 'Vec2', 'Vec3', 'Vec4', 'Vec5']))
```

```python
cols = ["IR %", "log(NSE)", "log(m3)", "log(iip)", "log(cop)"]
beta = pd.DataFrame(coin.evec[:, :2], index=cols, columns=['Eq1', 'Eq2'])

# Normalize each equation so that NIFTY coefficient = 1
for col in beta.columns:
    beta[col] = beta[col] / beta.loc["log(NSE)", col]

print(beta)
```

```python
new_india_data.head()
```

```python
data_np = new_india_data.iloc[:,1:].values
beta = coin.evec[:, :2]   # r = rank (number of cointegrating equations)

means = data_np.mean(axis=0)
intercept = -means @ beta

intercept_df = pd.Series(intercept, index=[f'Eq{i+1}' for i in range(beta.shape[1])])
print("Approximate intercepts for each cointegrating equation:\n", intercept_df)
```
